```python
import pandas as pd
```

```python
btw = pd.read_csv('https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/bwt.csv')
```

```python
primer_hijo = []
segundo_hijo = []
for i in range(len(btw['parity'])):
    if btw['parity'][i] == 0:
        segundo_hijo.append(btw['gestation'][i])
    else:
        primer_hijo.append(btw['gestation'][i])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(primer_hijo,label='Primer hijo')
plt.plot(segundo_hijo, label='Segundo hijo') 
plt.ylabel('Semanas de gestaci√≥n')  
plt.legend()
plt.show()
```
